```python
import GetOldTweets3 as got
from datetime import date, timedelta
from sqlite3 import connect
import pandas as pd
import time
import os
```

```python
df_Cities=pd.read_excel('uscities.xlsx')
```

```python
df_Cities=df_Cities[df_Cities['population']>75000]
print(df_Cities.shape[0])
len(df_Cities['state_id'].unique())
```

```python
L=['covid','covid-19' , 'covid19' ,'pandemic','reopen','shutdownorder','lockdownorder','stayathomeorder']

searchKeywords ='%20OR%20'.join(L[:6])
print(searchKeywords)
startDate = date(2020, 3, 1)+timedelta(days=0)
endDate =startDate+timedelta(days=76)
MaxTweets=10000
Within="10mi"
Set_Emoji="unicode"
Sleep_time=16*60
print(startDate)
print(endDate)
```

```python
len ([state for state in df_Cities['state_id'].unique() if not os.path.exists(state+'Tweet.xlsx')])
```

```python
# j=6, 9, 18, 19, 20, 29
j=0
#for state in df_Cities['state_id'].unique():
for state in [state for state in df_Cities['state_id'].unique() if not os.path.exists(state+'Tweet.xlsx')]:
    
    print(20*'--','j=',j,20*'--')

    df=pd.DataFrame()
    writer = pd.ExcelWriter(state+'Tweet.xlsx')

    for i in df_Cities[df_Cities['state_id']==state].index:
        print(state,'\t', df.shape[0],'\t', df_Cities.loc[i,'city'])


        for d in range(0,76,5):


            startDate = date(2020, 3, 1)+timedelta(days=d)
            endDate =startDate+timedelta(days=5)
            print(startDate)
            #print(endDate)
            city=str(df_Cities.loc[i,'lat'])+','+str(df_Cities.loc[i,'lng'])



            tweetCriteria = got.manager.TweetCriteria().setQuerySearch(searchKeywords)\
                                                           .setSince(str(startDate))\
                                                           .setUntil(str(endDate))\
                                                           .setNear(city)\
                                                           .setMaxTweets(MaxTweets)\
                                                           .setWithin(Within)\
                                                           .setEmoji(Set_Emoji)
            k=0
            while k<5:
                #print('\t','try=',k)
                try: 
                    for tweet in got.manager.TweetManager.getTweets(tweetCriteria):
                        x={'pri_id': tweet.id,
                        'text': tweet.text,
                        'CreateDateTime': tweet.date.strftime("%Y-%m-%d %H:%M"),
                        'CreateDate': tweet.date.strftime("%Y-%m-%d"),
                        'geo': tweet.geo,
                        'to_p': tweet.to,
                        'permalink': tweet.permalink,
                        'username':tweet.username,
                        'retweets': tweet.retweets,
                        'favorites':tweet.favorites,
                        'hashtags' :tweet.hashtags,
                        'mentions' :tweet.mentions,
                        'lat': df_Cities.loc[i,'lat'],
                        'lng': df_Cities.loc[i,'lng'],
                        'city': df_Cities.loc[i,'city'],
                        'state_id': df_Cities.loc[i,'state_id'],
                        'state_name': df_Cities.loc[i,'state_name'],
                        'Loc_id':df_Cities.loc[i,'id']}

                        df=df.append(x,ignore_index=True)
                    k=5
                except:
                    print('Error','\t','try=',k,'  \t', startDate)
                    k+=1
                    time.sleep(Sleep_time)
                #print(x)

    df.to_excel(writer,'Sheet1')
    writer.save()
    j+=1 
```
